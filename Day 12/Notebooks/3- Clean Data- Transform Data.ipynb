```python
import os
import pandas as pd
```

```python
cwd = os.getcwd() #to grab dirname of absolute path of (__file__)
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data')
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
working_file= os.path.join(CACHE_DIR, 'movies-box-office-dataset.csv')
output_file= os.path.join(CACHE_DIR, 'movies-box-office-dataset_cleaned.csv')
```

```python
df= pd.read_csv(working_file)
```

```python
df.head(120)
```

```python
#to remember to set this column
df['Rank']= -1
```

```python
df.head()
```

```python
df['Domestic %'] = df['%']
df['Foreign %'] = df['%.1']
df.head()
```

```python
#df.drop(columns=['%', '%.1']) 1 way to write it or
#new_df= df.drop(['%', '%.1'], axis=1,inplace=True) or
df.drop(['%', '%.1'], axis=1, inplace=True)
```

```python
df.head()
```

```python
#an attempt at fropping th dollar sign ij worldwide column
df['Worldwide']= df['Worldwide'].replace('$', '').replace(',', '')
```

```python
df.head()
```

```python
#another attempt at dropping the dollar sign ij worldwide column
to_clean_cols = ['Worldwide', 'Domestic', 'Foreign']

def currency_str_to_int(current_val):
    currency_val = current_val.replace('$', '').replace(',', '')
    try:
        currency_val = int(currency_val)
    except:
        # Takes any row value with "-" and turns into 0
        currency_val = 0
    return currency_val


def clean_col(row):
    
    #row = Pandas Series
    #dataframe = collecrtion of pandas series
   
    #print(row)
    for col in to_clean_cols:
        current_val = row[col]
        row[col] = currency_str_to_int(current_val)
    return row
    #pass
df_cleaned = df.apply(clean_col, axis=1)
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df.dtypes
```

```python
df_cleaned.sort_values(by=['Worldwide'], inplace=True, ascending=False)
df_cleaned.head()
```

```python
df_cleaned.reset_index(inplace=True, drop=True)
df_cleaned.head()
```

```python
df_cleaned['Rank'] = df_cleaned.index + 1
df_cleaned.head()
```

```python
df_cleaned ['Domestic %'] = df_cleaned ['Domestic'] / df_cleaned['Worldwide']
df_cleaned ['Foreign %'] = df_cleaned ['Foreign'] / df_cleaned['Worldwide']
```

```python
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.to_csv(output_file, index= False)
```
