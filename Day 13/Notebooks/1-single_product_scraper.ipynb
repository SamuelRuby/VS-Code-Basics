```python
url = 'https://www.amazon.com/Fitlaya-Fitness-Equipment-Adjustable-Equipment/dp/B0818T77J3/ref=sr_1_10?qid=1652624575&rnid=3408271&s=sporting-goods&sr=1-10'
title_lookup = '#productTitle' 
price_lookup = '#a-price a-text-price'
#price_lookup = "#priceblock_ourprice"
```

```python
#requests not working
r = requests.get(url)
html_str = r.text
```

```python
#print(html_str)
driver.get(url)
body_el = driver.find_element_by_css_selector('body')
html_str = body_el.get_attribute('innerHTML')
print(html_str)
```

```python
html_obj = HTML(html=html_str)
```

```python
product_title = html_obj.find(title_lookup, first=True).text
print(product_title)
```

```python
product_price=html_obj.find(price_lookup, first=True)
print(product_price)
```
